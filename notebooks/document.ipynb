# LangChain Chunking

```python
from langchain_community.document_loaders import DirectoryLoader, TextLoader


dir_text_loader = DirectoryLoader(
    "../data/text_data/",
    glob="**/*.txt",
    loader_cls=TextLoader,
    loader_kwargs={"encoding":'utf-8'}
)

resume_content = dir_text_loader.load()
resume_content
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

def split_documents_chunk(documents, chunk_size=200, chunk_overlap=200):
    splitter = RecursiveCharacterTextSplitter(
        chunk_size=chunk_size,
        chunk_overlap=chunk_overlap,
        length_function=len, 
        separators=['\n\n', "  "]
    )

    split_doc = splitter.split_documents(documents=documents)
    print(f"Splitter document size {len(documents)} into {len(split_doc)} chunks")

    if split_doc:
        print("Example of chunks ")
        print("Content: ", split_doc[0].page_content[:200])
        print("Metadata: ", split_doc[0].metadata)

    
    return split_doc

```

```python
chunks = split_documents_chunk(documents=resume_content)
chunks
```

# LangChain Embeddings
- sentence-transformers
- faiss-cpu
- chromadb


```python
import chromadb
import numpy as np
from sentence_transformers import SentenceTransformer
from chromadb.config import Settings
import uuid
from typing import Any, Dict, List, Tuple
from sklearn.metrics.pairwise import cosine_similarity
```

```python
embedding_model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
# 2️⃣ Initialize ChromaDB client
client = chromadb.Client(
    Settings(
        persist_directory="./chroma_db",
        anonymized_telemetry=False
    )
)
```

```python
# 3️⃣ Create / get collection
collection = client.get_or_create_collection(name="my_embeddings")
```

```python
# 5️⃣ Generate embeddings
ids = []
document_page_contents = []
embeddings = []

for text in chunks:
    ids.append(str(uuid.uuid4()))
    document_page_contents.append(text.page_content)
    embeddings.append(embedding_model.encode(str(text.page_content)))

collection.add(
    ids=ids,
    embeddings=embeddings,
    documents=document_page_contents
)
```

```python
collection
```

```python

# 7️⃣ Query example
query = "Which tools does Vijay use for MLOps and CI/CD?"
query_embedding = embedding_model.encode([query]).tolist()

results = collection.query(
    query_embeddings=query_embedding,
    n_results=10
)

print("Query Results:", results)
```
